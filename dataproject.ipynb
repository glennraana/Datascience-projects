```python
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
player = "Roger Federer" 
data = pd.read_csv('https://github.com/ipython-books/'
                 'cookbook-2nd-data/blob/master/'
                 'federer.csv?raw=true',
                 parse_dates=['start date'],
                 dayfirst=True)
data.head()
```

```python
# We want to add a collum if player had won or not 

data["win"] = data["winner"] == player
data["win"].tail()
```

```python
data.head()
```

```python
# making a variabel named won where we calculate the persent of rogers win 
won = 100*data["win"].mean()
print (f"{player} has won {won:.0f}% of his matches")
```
