```python
#inport what we need
import pandas as pd
import requests
import json
```

```python
#Welcome message and input variable
print("Welcome to the Weather planer")
location = input("Where are you going?:")
```

```python
# The API call to get weather on location
import requests

url = f"https://forecast9.p.rapidapi.com/rapidapi/forecast/{location}/summary/"

headers = {
	"X-RapidAPI-Key": "xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx",
	"X-RapidAPI-Host": "forecast9.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

print(response.json())
json_data = response.json()
```

```python
# Find out wich keys we have to work with
keys = json_data.keys()
print("Keys in the JSON data:")
for key in keys:
    print(key)
```

```python
# Want to look at the key forecat to get to know how to slice them to get the data i want.
forecast_location = json_data["forecast"]["items"]
print(forecast_location)
```

```python
# SLicing out the responses and make our app answer.
first_location = json_data["location"]["name"]
first_item_in_list = json_data["forecast"]["items"][0]
min_temp = first_item_in_list["temperature"]["min"]
max_temp = first_item_in_list['temperature']['max']
sun_hours = first_item_in_list["sunHours"]
rain_hours = first_item_in_list["rainHours"]

print(f"Today the lowest teprature in {first_location} would be {min_temp}")
print(f"Today the Highest teprature in {first_location} would be {max_temp}")
print(f"Today it wil be a total of {sun_hours} Sun Houers in {first_location}")
print(f"Today it wil be a total of {rain_hours} Rain Houers in {first_location}")

```
